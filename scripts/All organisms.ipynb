## Organisms list
These scripts aim at saving all organisms data from the KEGG api, the main goal is to create a data frame with alll of this information for it´s later analysis or to complete any dataframe with missing info of the organisms

```python
import requests
import re 
import pandas as pd
```

The function `get_org_data()` retrieves a GET petition and creates a list from it, that list contains (allegedly) all the organisms in the KEGG database, each line containing:  
1. Entry
2. 3 letter code
3. Organism
4. Phylogeny

The function `data_to_cols()` utilizes `data_org` as a parameter and distributes the content into multiple lists

```python
def data_to_cols(data_org):
    entry, code, organism, phylogeny = [], [], [], []
    for line in data_org:
        current_line = line.split('\t')
        entry.append(current_line[0])
        code.append(current_line[1])
        organism.append(current_line[2])
        phylogeny.append(current_line[3])
    return entry, code, organism, phylogeny
entry, code, organism, phylogeny = data_to_cols(data_org)
```

The function `list_to_dic()` uses the multiple lists created by `data_to_cols()` to create a dictionary for later transformation into a dataframe using $\text{Pandas}$ 

```python
def list_to_dic(entry, code, organism, phylogeny):
    data_dic =	{
      "entry": entry,
      "code": code,
      "organism": organism,
        "phylogeny": phylogeny}
    return data_dic
```
